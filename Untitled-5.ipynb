```python
list1=[]
list2=[]
n1=int(input("enter the length of list1:"))
n2=int(input("enter the length of list2:"))
print("enter the element of list1:")
for i in range(n1):
    m=input()
    list1.append(m)
print("enter the element  of list2:")
for i in range(n2):
    n=input()
    list2.append(n)
dic={list1[i] :list2[i] for i in range(min(len(list1),len(list2)))}
print(dic)
```
